```python
# 1. data visualization: 
# variable -> multi-variate modelling 
# 2. modelling -> prediction -> objective: to maximize the accuracy 
# 1. dataset has deficts: data preprocessing: 1. data imputation 2. hetergeneous data type fusion 3. .... 
# model building

# label encoding of revenue
from sklearn.preprocessing import LabelEncoder
data = data1
le = LabelEncoder()
data['Revenue'] = le.fit_transform(data['Revenue'])
data['Revenue'].value_counts()
```

```python
# getting dependent and independent variables

x = data
# removing the target column revenue from x
x = x.drop(['Revenue'], axis = 1)

y = data['Revenue']
count = 0
for i in range(len(y)):
    if y[i] == 1:
        count = count + 1
print (count)
# checking the shapes
print("Shape of x:", x.shape)
print("Shape of y:", y.shape)
# muli-variate regresion model 
# y = w1* x*1 + w2* x2 + ... + w28 * x28
```

```python
# spliting data set into: train and test (train: 2/3 test: 1/3)
from sklearn.model_selection import train_test_split
import sklearn
x_train, x_test, y_train, y_test = train_test_split(x, y, test_size = 0.333, random_state = 0)

# checking the shapes

print("Shape of x_train :", x_train.shape)
print("Shape of y_train :", y_train.shape)
print("Shape of x_test :", x_test.shape)
print("Shape of y_test :", y_test.shape)
```

```python
# 1. model : Descison Tree () -> entropy
from sklearn.ensemble import RandomForestClassifier
# build random forest object
# default parameters -> parameter tunning 
model_forest = RandomForestClassifier()
print(model_forest)
```

```python
# train model 
model_forest.fit(x_train, y_train)
```

```python
# this the model predict result 
y_predict = model_forest.predict(x_test)
from sklearn.metrics import confusion_matrix
from sklearn.metrics import classification_report
# evaluating the model
print("Training Accuracy :", model_forest.score(x_train, y_train))
print("Testing Accuracy :", model_forest.score(x_test, y_test))

# Homework (optional): confusion matrix 
# confusion matrix
cm = confusion_matrix(y_test, y_predict)
plt.rcParams['figure.figsize'] = (6, 6)
sns.heatmap(cm ,annot = True)

# classification report
cr = classification_report(y_test, y_predict)
print(cr)
```

```python
# feature selection to improve model performance
from sklearn.feature_selection import VarianceThreshold
sel = VarianceThreshold(threshold = (0.7 * (1 - 0.70)))
sel.fit_transform(x)
x_train, x_test, y_train, y_test = train_test_split(x, y, test_size = 0.333, random_state = 0)
# train model 
model_forest.fit(x_train, y_train)
# this the model predict result 
y_predict = model_forest.predict(x_test)
from sklearn.metrics import confusion_matrix
from sklearn.metrics import classification_report
# evaluating the model
print("Training Accuracy :", model_forest.score(x_train, y_train))
print("Testing Accuracy :", model_forest.score(x_test, y_test))

# Homework (optional): confusion matrix 
# confusion matrix
# TP, FN, FP, TN
# accuracy = (TP + TN)/(TP + FN + FP + TN)
# precision = (TP) / (TP + FP) 
# recall = TP /(TP + FN)
# F1_score = 2 * TP/ (2* TP + FP + FN)
# evaluation metrics 
cm = confusion_matrix(y_test, y_predict)
plt.rcParams['figure.figsize'] = (6, 6)
sns.heatmap(cm ,annot = True)

# classification report
cr = classification_report(y_test, y_predict)
print(cr)
# feature selection: improve model's accuracy scores or to boost their performance on very high-dimensional dataset. 
#1. removing features with low variance (Variance Threshold): a simple baseline approach for feature selection. It removes all features whose variance 
# does not meet some threshold. By default, it removes all zero-variance features. Var[X] = p(1-p) (p is a parameter). Assume p =. 0.8, threshold = 0.16 
# Univariate feature selection
# recursive feature elimination
```

```python
# what is model?
# model = parameters + mathematical equations 
# parameters: numeric values 
# mathematical equations: 
# y = ax + b 
# logistic regression: 
# 1. regression 2. classfication 3. clustering 
# n input: x_1, x_2, x_3, ...., x_n
# one output: y 
# Step 1: y_t = a_1 * x_1 + a_2 * x_2 + a_3 * x_3 + ... + a_n * x_n 
# Step 2: y = 1/ (e^(-1 * y_t) + 1) -> non-linear transformation -> sigmoid function -> (deep neural network()) optional 


from sklearn.linear_model import LogisticRegression
model_logistic = LogisticRegression(random_state = 0)
sel = VarianceThreshold(threshold = (0.7 * (1 - 0.70)))
sel.fit_transform(x)
x_train, x_test, y_train, y_test = train_test_split(x, y, test_size = 0.333, random_state = 0)
model_logistic.fit(x_train, y_train) # train logistic regression model 
# this the model predict result 
y_predict = model_logistic.predict(x_test)
print (y_predict)
y_predict_prob = model_logistic.predict_proba(x_test)
print (y_predict_prob)
from sklearn.metrics import confusion_matrix
from sklearn.metrics import classification_report
# evaluating the model
print("Training Accuracy :", model_logistic.score(x_train, y_train))
print("Testing Accuracy :", model_logistic.score(x_test, y_test))

# Homework (optional): confusion matrix 
# confusion matrix
cm = confusion_matrix(y_test, y_predict)
plt.rcParams['figure.figsize'] = (6, 6)
sns.heatmap(cm ,annot = True)

# classification report
cr = classification_report(y_test, y_predict)
print(cr)
```

```python
# SVM: support vector machine ()
# support vector: the points that are the closest to the classifier lines 
# objective: adding the distance of all support vectors until it is minimized. When the total distance is minized, then we find the optimal line to 
#successfully classify the two kinds of points (samples)
# kernel function 
from sklearn import svm
s_v_m = svm.SVC(probability=True)
sel = VarianceThreshold(threshold = (0.7 * (1 - 0.70)))
sel.fit_transform(x)
x_train, x_test, y_train, y_test = train_test_split(x, y, test_size = 0.333, random_state = 0)
s_v_m.fit(x_train, y_train)
y_predict = s_v_m.predict(x_test)
y_predic_prob = s_v_m.predict_proba(x_test)
print(y_predic_prob)
from sklearn.metrics import confusion_matrix
from sklearn.metrics import classification_report
# evaluating the model
print("Training Accuracy :", s_v_m.score(x_train, y_train))
print("Testing Accuracy :", s_v_m.score(x_test, y_test))

# Homework (optional): confusion matrix 
# confusion matrix
cm = confusion_matrix(y_test, y_predict)
plt.rcParams['figure.figsize'] = (6, 6)
sns.heatmap(cm ,annot = True)

# classification report
cr = classification_report(y_test, y_predict)
print(cr)
```
